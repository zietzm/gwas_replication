```python
import pathlib

import pandas as pd
```

```python
gwas_files = list(pathlib.Path('../data/gwas_catalog/').glob('*.tsv'))

gwas_df = pd.concat(pd.read_csv(path, sep='\t') for path in gwas_files)

gwas_df.to_csv('../data/gwas_catalog/combined.csv', index=False)

print(gwas_df.columns)

gwas_df.head(2)
```
